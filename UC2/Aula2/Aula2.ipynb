```python
import numpy as np

dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
print(dados)
display(dados)

q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)

print(f"Primeiro quartil (Q1): {q1}")
print(f"Segundo quartil (Q2): {q2}")
print(f"Terceiro quartil (Q3): {q3}")
```

## Breve introdução a graficos com o .plot()

```python
import pandas as pd 
import matplotlib.pyplot as plt

arquivo = 'base_invest.xlsx'
df_transacoes = pd.read_excel(arquivo)

contagem_operacao = df_transacoes['operacao'].value_counts()

contagem_operacao.plot(kind='barh', title='Tipos de Operação')

plt.show()
```
